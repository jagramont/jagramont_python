```python
#TEST THE IRIS DATASET

import seaborn as sns
iris = sns.load_dataset("iris") #the quotations marks must be the correct ones
iris.head()
```

```python
# Convert example dataset “iris” to a dataframe in pandas
df = pd.DataFrame(iris)
```

```python
# A series on pandas could be seen as a column of dataframe
s = df["sepal_length"] 
se = pd.Series(s)
se
```

```python
# Add a column to dataframe:
df["test_column"] = "test"
df
```

```python
# Delete a column
del df["test_column"]
df
```

```python
# Add a column with new values calculated from operating other columns
df["added_length"] = df["sepal_length"]+df["petal_length"]
df
```

```python
# STATISTICS
# get the mean
df["sepal_length"].mean()
```

```python
# get aggregates statistics
df["sepal_length"].describe()
```

```python
# get aggregates statistics for several columns
df[["sepal_length","sepal_width", "petal_length", "petal_width"]].describe()
```

```python
# get aggregates statistics for several columns and group by species
df[["sepal_length","sepal_width", "petal_length", "petal_width"]].groupby(df["species"]).describe()
```
